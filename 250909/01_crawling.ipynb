```python
# íŠ¹ì • ì£¼ì†Œì— ìš”ì²­ì„ ë³´ë‚¸ë‹¤.
res = requests.get('https://finance.naver.com/')
# resì—ì„œ html ë°ì´í„°ë¥¼ ë¬¸ìì—´ë¡œ ì €ì¥
html_data = res.text
# BeautifulSoupì„ ì´ìš©í•˜ì—¬ ë°ì´í„° íŒŒì‹±
soup = bs(html_data, 'html.parser')
```

```python
# soupì—ì„œ div íƒœê·¸ ì¤‘ classì˜ ê°’ì´ 'section_sise_top' íƒœê·¸ë¥¼ ì°¾ëŠ”ë‹¤.
# í•´ë‹¹ ì¡°ê±´ì˜ íƒœê·¸ê°€
len(
    soup.find_all(
        'div',
        attrs= {
            'class' : 'section_sise_top'
        }
    )
)
```
