#### 파생변수
- 기존의 데이터를 이용하여 새로운 데이터 생성
- 기존의 데이터만으로 설명이 불가능하거나 상관관계가 약한 경우,</br>
새로운 데이터를 생성해 설명이 가능하고 상관관계가 강해지도록 함
    - case1: 반복문(for문, while문, map(  )) 사용
    - case2: 함수 만들어 사용

```python
# 총확진자수 데이터를 이용해 일일확진자 데이터 생성
# 일일확진자 = 오늘의 총확진자수 - 어제의 총확진자수
# df.head()를 보면, 2020-03-10 이전의 데이터는 없으므로 그 다음 인덱스부터 확인 가능
df.loc[ 1, '총확진자'] - df.loc[ 0, '총확진자']
```

case1 - for문

```python
# 반복문을 이용하여 일일확진자수를 리스트로 생성해 데이터프레임에 대입
# 0번째는 2020-03-10의 데이터이므로 초기값이 들어가있어야 함.
decide_list = [0]

# 반복문 시작(데이터프레임의 길이에 -1 반복 실행)
# for i in range(0, len(df)-1): 또는 ↓
for i in range(1, len(df)):
    # i = 1 부터 819 대입하여 반복
    decide_cnt = df.loc[ i, '총확진자'] - df.loc[ i-1, '총확진자']
    decide_list.append(decide_cnt)

# df에 일일확진자 컬럼 생성하고 데이터로 decide_list 대입
    # ⇒ dict에서 key:value를 추가하는 형태와 같다.
df['일일확진자'] = decide_list
```

case2 - 함수

<span style='color:#ffd33d'>shift( )
- 인덱스를 이동하는 함수 shift(  n  )
    - n만큼 인덱스 변경<span style='color:#808080'>(n의 기본값= 1)</span>
> 인덱스 이동</br>
  -1  : 한 칸 위로</br>
  +1  : 한 칸 아래로

```python
# 인덱스를 이동하는 함수 shift(n) -> n만큼 인덱스 변경(n의 기본값= 1)
df['총사망자'].shift()
# 데이터 타입 float
```

```python
# shift()한 데이터의 첫 번째 값이 결측치 -> 0번째 데이터 NaN
# Series이므로 굳이 list에 넣지 않고 바로 데이터프레임에 넣을 수 있음
df['일일사망자'] = df['총사망자'] - df['총사망자'].shift()
```

<span style='color:#ffd33d'>diff( )
- 특정 시리즈 안에서의 차이를 구하는 함수 diff(  n  )
    - n은 인덱스 수<span style='color:#808080'>(n의 기본값= 1)</span>

```python
df['총사망자'].diff()
```

```python
# 일일사망자, 일일확진자 데이터는 범위가 0보다 크거나 같아야 한다.
# 비교연산자를 사용해 해당 데이터를 확인한다.
df['일일사망자'] >= 0

# 0번 인덱스에는 결측치가 들어있기 때문에 False
```

```python
# 따라서 일일사망자 컬럼의 데이터에서 결측치는 0으로 채워준다.
df['일일사망자'].fillna(0, inplace=True)

# FutureWarning: 코딩 하다보면 여러 warning 메시지가 많이 나오지만 신경 안 써도 된다.
```

```python
# 다시 비교연산자를 사용해 해당 데이터를 확인한다.
df['일일사망자'] < 0
```

```python
# 인덱스의 조건식을 활용해
# 일일사망자가 음수인 데이터 확인
flag = df['일일사망자'] < 0
df.loc[flag, ]
```

5. '일일사망자' 컬럼은 diff(  ) 함수를 이용하여 생성

```python
# 1. '등록일시'를 기준으로 오름차순 정렬
df.sort_values(
    ['등록일시'], ascending=True, inplace=True
)
```

```python
df.loc[ 444:448, ]
```

```python
# 2. 인덱스를 초기화하고 기존의 인덱스 제거
df.reset_index(drop=True, inplace=True)
# df.reset_index().drop('index', axis=1)
```

```python
# 3. '일일확진자', '일일사망자' 컬럼 제거
df.drop(
    ['일일확진자', '일일사망자'],
    axis=1,  # axis='columns'
    inplace=True
)
```

```python
# 4. '일일확진자' 컬럼은 shift() 함수를 이용하여 생성
# 파생변수
# df['총확진자'] - df['총확진자'].shift()    # shift 기본값 1(한 칸씩 아래로 내림)
# 0번째 인덱스 Nan

# df['총확진자'] - df['총확진자'].shift().fillna()  --> X
# shift한 값에 결측치를 채우려는 게 아니라, 그 차에 결측치를 채우려는 것이므로 () 쳐주기.

df['일일확진자'] = (df['총확진자'] - df['총확진자'].shift()).fillna(0)
```

```python
# 5. '일일사망자' 컬럼은 diff() 함수를 이용하여 생성
# 파생변수
df['일일사망자'] = df['총사망자'].diff().fillna(0)
```

```python
# 6. '일일확진자', '일일사망자' 컬럼의 데이터에서 음수가 존재하는지 확인
flag = (df[ ['일일확진자', '일일사망자'] ] < 0).any(axis=1)
df.loc[flag,]
```
