### 은행 계좌 class 생성
- class 변수 생성 (동일한 class를 생성했을 때 서로 공유하는 변수)
    - total_cost = 0     # 총 금액
    - user_cnt = 0       # 유저 수
- 생성자 함수 (class 간 독립적인 변수 생성)
    - 유저의 이름 : name         *(str)*
    - 유저의 생년월일 : birth     *(str)*
    - 계좌 잔액 : cost           *(int)*
    - 계좌 기록 : log            *(list)*
    - 유저의 수(user_cnt) 1 증가
</br>-----------중간점검 1-----------
</br>유저의 이름, 생년월일, 계좌 잔액은 단일데이터
</br>계좌 기록은 자료형 데이터 (tuple은 수정 불가하므로 list 혹은 dict)
- 함수 생성
    - **입/출금 함수**
        - 매개변수 *(바뀌는 부분)*
            - 입금과 출금을 확인하는 변수 (_type) : 입금 0, 출금 1
            - 금액 (_cost)
        - _type이 0이라면 계좌의 잔액(독립적인 변수) 증가</br>& 은행이 가진 총액(class 변수)을 증가시킨다.
        - _type이 1이라면 계좌의 잔액(독립적인 변수) 감소</br>& 은행이 가진 총액(class 변수)을 감소시킨다.
            - _cost가 self.cost보다 작거나 같은 경우
                - 계좌의 잔액 감소, 은행의 총액 감소  -> 출금해준 것
            - 위의 조건이 거짓이라면
                - 계좌는 그대로 유지
                - '잔액 부족' 메시지 출력
        - 입/출금 내역을 self.log에 추가 (append())
</br>-----------중간점검 2-----------
    - **계좌의 입/출금 내역을 보여주는 함수**
        - 매개변수
            - 입금 / 출금 / 전체 내역을 확인할 수 있는 변수 (_mode)
                - 기본값은 9(전체)
                - 입금 내역은 0, 출금 내역은 1
        - _mode에 따라서 self.log에 있는 데이터를 필터해서 되돌려준다.
        </br>(또는 print로 출력)


```python
# class 선언
class Bank:
    # class 변수 선언
    total_cost = 0  # 은행 총액 # 유저가 입금 시 증가, 출금 시 감소
    user_cnt = 0    # 유저 수   # class 생성될 때(유저가 계좌를 생성할 때) 1씩 증가
    
    # 생성자 함수(__init__) 선언
    def __init__(self, _name, _birth):  # 계좌 생성 시 잔액과 기록 없으므로 이름, 생년월일만 매개변수로 받음
        # 독립적인 변수를 선언
        self.name = _name
        self.birth = _birth
        self.cost = 0
        self.log = []
        # 유저의 수(class 변수)를 1 증가시킨다.
        Bank.user_cnt += 1
    # -----------중간점검 1-----------

    # 입/출금 함수 선언
    def change_cost(self, _type, _cost):
        # _type이 0이라면 입금
        if _type == 0:
            # self.cost에 _cost만큼 증가
            self.cost += _cost
            # Bank.total_cost에 _cost만큼 증가
            Bank.total_cost += _cost
            # log를 추가 -> dict 형태의 데이터를 추가
            dict_data = {
                '타입' : '입금',
                '금액' : _cost,
                '잔액' : self.cost
            }
            # self.log에 dict_data를 추가 -> self.log는 2차원 데이터 생성
            self.log.append(dict_data)
            print(f"입금완료 : 잔액은 {self.cost}입니다.")
        # _type이 1이라면 출금
        elif _type == 1:
            # self_cost가 _cost보다 크거나 같은 경우
            if self.cost >= _cost:
                # 출금 가능
                # self.cost를 _cost만큼 감소
                self.cost -= _cost
                # Bank.total_cost를 _cost만큼 감소
                Bank.total_cost -= _cost
                # dict_data 생성 -> self.log에 추가
                dict_data = {
                    '타입' : '출금',
                    '금액' : _cost,
                    '잔액' : self.cost
                }
                self.log.append(dict_data)
                print(f"출금완료 : 잔액은 {self.cost}입니다.")
        else:
            print('현재 잔액이 부족합니다.')
            # 로그에 추가하지는 않음 (단지 잔액 부족 기록이 보기 좋지 않아서)
    # -----------중간점검 2-----------
    
    def view_log(self, _mode = 9):
        # 입/출금 내역(self.log) 출력
        # _mode 기본값 9(전체 내역 출력)
        if _mode == 9:
            # self.log를 기준으로 반복문 생성
            for log_data in self.log:
                print(log_data)
        # _mode가 0이면 입금 내역만 출력
        elif _mode == 0 :
            for log_data in self.log:
                # log_data의 타입-> dict{'타입': 'xx', '금액':'xx', '잔액':'xx'}
                # log_data에서 key가 '타입'인 value의 값이 '입금'이라면
                if log_data['타입'] == '입금':
                    print(log_data)
        # _mode가 1이면 출금 내역만 출력
        elif _mode == 1:
            for log_data in self.log:
                if log_data['타입'] == '출금':
                    print(log_data)
        else:
            print("_mode의 값이 잘못되었습니다.")
    # -----------중간점검 3-----------
```

### class의 상속
- class 간의 상속은 부모 class에 있는 기능(변수, 함수)을 자식 class에서 사용 가능하게 만드는 기능
- 부모 class의 기능은 그대로 유지한 상태에서 자식 class에서 해당 기능을 그대로 사용하거나 약간의 변형 가능
- 자식 class에서 부모 class에 접근하는 방법: super()

```python
# 부모 class 선언
class A:
    # 생성자 함수(class 안에서 독립적으로 사용 가능한 변수 생성)
    def __init__(self, _name, _birth):
        self.name = _name
        self.birth = _birth
    # class에 저장된 유저의 정보를 출력하는 함수
    def view_info(self):
        print(f"이름은 {self.name}이고 나이는 {self.birth}입니다.")
```

```python
test_user = A('kim', 20)
```

```python
test_user.view_info()
```

```python
# 자식 class 선언
# A class의 기능을 상속 받는다.
class B(A):       # 상속 받을 class가 있다면 class명 옆에 (상속할 class명)을 붙여준다. 없으면 생략.
    # B class는 A의 변수 2개와 view_info() 함수 사용 가능
    # 생성자 함수
    def __init__(self, _name, _birth, _loc):    # A class는 이름, 나이. B class는 이름, 나이, 지역을 받음.
        # 부모 class 에서 저장하는 부분
        # self.name = _name  <-
        # self.birth = _birth    <- super() 대신 이 두 줄을 넣어도 결과는 같다.
        # 간단하게 저장이 되지 않는 경우, 부모 class(super())에 있는(.) 생성자 함수(__init__()) 호출
        super().__init__(_name, _birth)
        self.loc = _loc
    def view_info(self):
        print(f"이름은 {self.name}이고 나이는 {self.birth}, 지역은 {self.loc}입니다.")

        # 부모 class에 있는 view_info() 호출
        super().view_info()
        # 추가 기능
        print(f"지역은 {self.loc}입니다.")
```

```python
test_user2 = B('park', 30, 'seoul')
```

```python
# test_user2에 사용 가능한 함수는 2개 -> view_info(), view_info2()
print(test_user2.view_info())
# 결과
# 이름은 park이고 나이는 30, 지역은 seoul입니다.
# None         <-- 함수에 return이 없기 때문에 None도 출력됨
```

### User class 선언
- Bank class의 기능을 상속받는다.
- 일의 종류와 금액을 나타내는 dict 형태의 데이터를 클래스 변수에 저장
- 구매할 상품의 종류와 금액을 나타내는 dict 형태의 데이터를 클래스 변수에 저장
- 생성자 함수 선언
    - Bank class 안에 변수 _name, _birth 데이터를 받는다.
    - 구매한 물건의 목록을 저장하기 위해 비어있는 리스트 생성
- 일을 한다 -> work() 함수 선언
    - 매개변수
        - 일의 종류(_type)
    - type에 따라서 금액 지정
        - dict를 이용해 금액 지정
    - 유저의 잔액을 금액만큼 증가
        - Bank *(부모 class)* 안에 있는 change_cost() 함수를 이용
- 지출, 구매하는 함수 -> buy()
    - 매개변수
        - 물건의 종류(_type)
    - _type에 따라서 금액 지정
        - dict를 이용하여 금액 지정
    - 구매하려는 물건의 금액과 유저의 잔액(self.cost) 비교
        - 유저의 잔액이 많거나 같다면
            - 부모 class의 change_cost() 함수 이용
            - 구매한 물건의 종류를 목록에 추가
        - 유저의 잔액이 적다면
            - "잔액이 부족합니다." 메시지 출력

```python
# User class 선언하고 Bank class의 기능을 상속 받는다.
class User(Bank):
    # class 변수 생성
    work_types = {
        "A" : 10000,     # "A"라는 일을 하면 1만원을 벌고,
        "B" : 20000,     # "B"라는 일을 하면 2만원을 벌고,
        "C" : 30000      # "C"라는 일을 하면 3만원을 번다.
    }
    item_list = {
        "학식": 6000,
        "돈까스": 7000,
        "햄버거": 8000,
        "탕수육": 19000
    }
    # 생성자 함수 만들어 변수 저장
    def __init__(self, _name, birth):
        # _name, _birth는 부모 class의 생성자 함수를 이용하여 저장
        super().__init__(_name, birth)
        # 유저가 구매한 물건의 목록을 생성하여 비어있는 리스트를 대입
        self.items = []
    
    # work() 함수 생성
    def work(self, _type):
        # type에 따라 금액이 지정되는데, work_types 딕셔너리에 저장되어 있다.
        # work_types에 없는 key를 지정하면 Error 발생하므로 예외 처리 필요
        try:
            # 실행할 코드 작성
            # 금액 지정
            cost = User.work_types[_type]
            # 잔액 증가
            # 부모 class에 있는 change_cost() 함수 호출
            super().change_cost(_type=0, _cost=cost)
        except:
            # try 영역에 있는 코드들이 실행중 문제가 발생했을 때
            print("work_types에 존재하지 않는 _type을 입력하였습니다.")
# -----------중간점검 1-----------

    # buy() 함수를 생성
    def buy(self, _type, _cnt = 1):    # 물건의 타입과 개수
        # _type에 따라 금액 지정
        try:
            # item_list에 있는 물건의 금액을 불러온다.
            cost = User.item_list[_type] * _cnt
            # 현재 잔액과 cost를 비교
            if self.cost >= cost:
                # 구매 성공
                # 부모 class에서 change_cost() 함수 호출
                super().change_cost(_type = 1, _cost = cost)
                # self.items에 구매한 물건과 개수 추가
                self.items.append(
                    f"{_type} x {_cnt}"
                )
            else:
                # 구매 실패(잔액 부족)
                print('구매 실패: 잔액이 부족합니다.')
        except:
            # 구매 실패
            # item_list 딕셔너리에 구매하려는 물건의 정보가 없을 때
            print('구매 실패: 구매하려는 물건의 정보가 존재하지 않습니다.')
    # -----------중간점검 2-----------

    # 유저의 정보를 출력해주는 함수 생성
    def user_info(self):
        # self에 저장된 변수들 중 name, birth, cost, log, items 중 log 제외하고 보여줄 것임
        # log는 Bank의 함수에서 확인하면 되니까
        print("""
            이름: {self.name}
            나이: {self.birth}
            잔액: {self.cost}
            구매한 물건의 목록: {self.items}
        """)     # """: 줄이 바뀌어도 하나의 텍스트로 인지

    # work_types, item_list 딕셔너리를 추가하는 함수 생성
    # self 제외 매개변수 3개 사용
    # 매개변수: 딕셔너리의 key(_key, str), value(_value, int),
    #          work_types를 바꿀건지 item_list를 추가할건지 선택(_select -> work 혹은 item))
    def add_type(self, _select, _key, _value):
        # _select가 "work"라면
        if _select == "work":
            # 클래스변수 work_types에 _key : _value 추가
            User.work_types[_key] = _value
        # _select가 "item"이라면
        elif _select == "item":
            # 클래스변수 item_list에 _key : _value를 추가
            User.item_list[_key] = _value
        # 그 외의 경우
        else:
            # "_select에는 work / item 만 사용이 가능합니다." 메시지 출력
            print("_select에는 work / item 만 사용이 가능합니다.")
    # -----------중간점검 3-----------
```
