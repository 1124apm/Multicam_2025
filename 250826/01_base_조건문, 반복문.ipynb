## 조건문

### 1. if문

```python
# if 조건문
# 참인 경우 실행하는 코드와 거짓인 경우 실행하는 코드를 작성
# 참 아니면 거짓인 경우의 코드 하나만 실행
# if 조건식(bool 데이터): 
#   실행할 코드(들여쓰기)
# else :
#   조건식이 거짓인 경우 실행할 코드(들여쓰기)

a = 5

if a < 10:
    # a가 10보다 작은 경우 실행이 되는 부분
    print("a는 10보다 작다")
else:
    # a가 10보다 크거나 같은 경우 실행이 되는 부분
    print("a는 10보다 크거나 같다")
```

```python
# 조건식이 여러 개인 경우
score = int(
    # input() : 실행할 때마다 데이터를 입력하는 창이 생성되어,
    # (문자 형태로) 입력값을 받아온다.
    input("점수를 입력하시오: ")
)
# print(score)

# score에 따라서 학점 출력
# 방법1. if문
# A: score가 90점 이상인 경우
if score >= 90:
    print('A')
# B: score가 90점 미만 80점 이상인 경우
if (score < 90) and (score >= 80):
    print('B')
# C: score가 80점 미만 70점 이상인 경우
if (score < 80) and (score >= 70):
    print('C')
# F: score가 70점 미만인 경우
if score < 70:
    print('F')
```

```python
# score에 따라서 학점 출력
# 방법2. if ~ else문
# A: score가 90점 이상인 경우
if score >= 90:
    print('A')
# score가 90점 미만인 경우
else:
    # B: score가 90점 미만 80점 이상인 경우
    if score >= 80:
        print('B')
    # score가 80점 미만인 경우
    else:
        # C: score가 80점 미만 70점 이상인 경우
        if score >= 70:
            print('C')
        # F: score가 70점 미만인 경우
        else:
            print('F')
```

```python
# score에 따라서 학점 출력
# 방법3. if ~ elif ~ else문
# A: score가 90점 이상인 경우
if score >= 90:
    print('A')
# B: score가 90점 미만 80점 이상인 경우
elif score >= 80:
    print('B')
# C: score가 80점 미만 70점 이상인 경우
elif score >= 70:
    print('C')
# F: score가 70점 미만인 경우
else:
    print('F')
```

### 3. while문

```python
# while문
# 초기값 지정 (시작값을 지정)
i = 1

while i <= 10:
    # i가 10보다 작거나 같은 경우에는 반복 실행
    print(i)
    # i의 값이 10보다 큰 경우가 거짓이므로
    # 조건식이 거짓이 될 수 있도록 i의 값이 증가하는 코드 작성
    i += 1
```

```python
# 1부터 10까지의 합계
i = 1
result = 0

while i <= 10:
    # result에 i를 더하고 result에 대입
    # 첫 번째 반복: i = 1, result = 0 + 1 = 1
    # 두 번째 반복: i = 2, result = 1 + 2 = 3
    result += i
    # i에 1을 더하고 i에 대입
    # 첫 번째 반복: i = 1 + 1 = 2, result = 1
    # 두 번째 반복: i = 2 + 1 = 3, result = 3
    i += 1
    # 반복 시행되는 코드가 없으므로 while문으로 되돌아간다.

print(result)
```

```python
# 1부터 10까지의 합계
i = 1
result = 0

while i <= 10:
    # i에 1을 더하고 i에 대입
    # 첫 번째 반복: i = 1 + 1 = 2, result = 0
    # 두 번째 반복: i = 2 + 1 = 3, result = 2
    i += 1
    # result에 i를 더하고 result에 대입
    # 첫 번째 반복: i = 2, result = 0 + 2 = 2
    # 두 번째 반복: i = 3, result = 2 + 3 = 5
    result += i
    print(result)
    # 반복 시행되는 코드가 없으므로 while문으로 되돌아간다.

print(result)

# 이렇게 하면 결과가 달라짐 (2~11의 합이 되므로)
```

### 4. break문

```python
# break문
# 반복문 강제 종료

for i in range(10000):
    print(i)
    break

print('반복문 종료')
```

```python
# 반복문 안에서 조건식을 이용하여 break를 이용
# 1부터 1씩 증가시킨 데이터의 누적합 값이 7,000이 넘어가는 시점은 언제인가?
# 즉, result가 7,000 이상이 되는 조건식 생성

# case1 (for문)
# 누적합 해줄 공간 생성
result = 0

for i in range(1, 1001, 1):
    # 위치1
    # result가 7,000 이상이면 반복문 종료
    if result >= 7000:
        break
    # i를 result에 누적합
    result += i
    # 위치2

print(result)
print(i)
```

```python
result = 0

for i in range(1, 120):
    result += i

print(result)

# 결과가 7140으로, if문이 위치1에 존재한 결과와 119만큼 차이난다
# 이 말은, 1부터 119이 아닌 118까지의 누적합이 7021이라는 것. 즉 i = 118.
```

```python
result = 0

for i in range(1, 1001, 1):
    # 위치1
    # i를 result에 누적합
    result += i
    # 위치2
    # result가 7,000 이상이면 반복문 종료
    if result >= 7000:
        break

print(result)
print(i)
```

```python
# while문을 이용하여 1부터 1씩 증가시킨 데이터의 누적합이 8,400을 넘어가는 시점을 구하라.

# 시작값 설정
i = 1

# 누적합의 초기값 설정
result = 0

# while 조건식
# 반복 횟수가 명확하지 않을 때 -> 무한 반복(조건식 True)
while True :
    # 위치1   
    # 시작값을 누적합에 더하고 대입
    result += i
    # 위치2
    # (누적합이 8,400 이상인 경우 반복문 종료)
    if result >= 8400:
        break
    # 시작값 증가
    i += 1
    # 위치3

# 시작값과 누적합의 값 출력
print(i, result)

# 위치1은 130회 반복, i = 131, result = 1~130 누적합
# 위치2는 129회 반복, i = 130, result = 1~130 누적합
# 위치3은 129회 반복, i = 131, result = 1~130 누적합
# 따라서 if문은 위치2가 올바른 위치임을 알 수 있음
# result에 대한 조건식은 result의 값이 변화하고 난 바로 뒤에 붙이는 게 맞을 듯?
```

### 5. continue문

```python
# continue문
# continue가 실행되면 아래의 반복 실행 코드를 스킵하고 반복문으로 되돌아간다.
# 반복 횟수는 증가!

for i in range(10):
    print(i)
    continue
    
```

```python
for i in range(10):
    continue
    print(i)
```

```python
for i in range(10):
    continue

print(i)
```

```python
# 1부터 100까지의 누적합 = 5050
result = 0
for i in range(1, 101):
    result += i

print(result)
```

```python
# 1부터 100까지의 짝수의 누적합은?
# 방법1. range 변경
result = 0
for i in range(2, 101, 2):
    result += i

print(result)
```

```python
# 1부터 100까지의 짝수의 누적합은?
# 방법2. i가 짝수일 때만 더하기
result = 0
for i in range(1, 101):
    if i % 2 == 0:
        result += i

print(result)
```

```python
# 1부터 100까지의 짝수의 누적합은?
# 방법3. i가 홀수일 때는 스킵하기
result = 0
for i in range(1, 101):
    if i % 2 == 1:
        continue
    result += i

print(result)
```
