```python
# 해당 api에서 사용할 서비스키 (함수에서 사용될 필수 매개변수의 인자값과 같다.)
# decoding 키
service_key = 'dtbWOdJ/Cz5HE0DGLU+CRPe7pOW0NIQBUcGEqsHZaTRiYCI/5+zugwzQjcvuId7NPdg6rUiW+ft3fm7yqyD4pw=='
# 해당 api의 주소 (python에서 호출할 함수의 이름과 같다.)
url = 'http://apis.data.go.kr/1262000/CountryPopulationService2/getCountryPopulationList2'

# 매개변수들을 하나의 변수에 dict 형태로 저장
# 변수 이름은 공공데이터에 이미 지정되어있음
_params = {
    'ServiceKey' : service_key,
    'pageNo' : 1,
    'numOfRows' : 10,
    'returnType' : 'JSON'
}

res = requests.get( url, params= _params)
res
```

```python
type(res.content)
```

```python
import json
```

```python
res_data = json.loads(res.content)
res_data
```

```python
# 선택사항
# 들여쓰기를 통해 라인별로 보여줌
# pprint = pretty + print
from pprint import pprint
```

```python
pprint(res_data)
```

```python
pprint(res_data['response']['body']['items']['item'])
# => 원하는 2차원의 데이터가 만들어짐
# => 데이터프레임을 만들거나, 데이터베이스 서버에 넣을 수도 있다!
```

```python
import pandas as pd
```

```python
# pandas의 DataFrame을 이용하여 res_data를 변경
df = pd.DataFrame(
    res_data['response']['body']['items']['item']
)
df.head()
```

```python
# 데이터프레임 타입 데이터의 타입을 변경
# DataFrame 안에 타입을 바꿔주는 함수 존재 -> 타입 변경 키워드(to) + 타입명
# 언제나 해당하는 파일이나 경로를 함께 써줘야 저장됨
df.to_dict('records')
```

```python
# 현재 경로에 test.csv 파일이 생성됨
# index 매개변수: index를 저장할 것인가? (기본값 True - 저장함)
df.to_csv('./test.csv', index= False)
```
