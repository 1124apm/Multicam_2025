## 자료형 데이터
1. 튜플(tuple)
- 여러개의 데이터를 하나의 변수에 저장시키기 위한 형태
- 자료형 데이터 중 가장 기본적인 형태
- ()를 이용해서 데이터 생성
- 한 번 생성되면 수정 불가능. 새로운 tuple을 생성해야 함

2. 리스트[list]
- []을 이용해서 데이터 생성
- tuple과는 다르게 각 원소의 수정, 삭제, 추가 가능(리스트의 내부함수 이용)
- 원소의 개수 0개부터 생성 가능 -> 여러 개의 데이터를 하나의 변수에 담을 준비 과정으로써 비어있는 리스트 생성

3. 딕셔너리{dict}
- {}을 이용하여 데이터 생성
- key : value 의 형태로 데이터 구성 (list에서는 위치를 기준으로 데이터를 추출한다면, dict는 key를 기준으로 데이터 추출)
- 원소의 개수는 0부터 생성 가능

### 1. 튜플

```python
# tuple 데이터 생성
# 원소마다 타입이 달라도 생성 가능
tuple_a = (1, 'Hello')
tuple_b = (3, 3.14)

print(tuple_a)
print(tuple_b)

print(type(tuple_a))
print(type(tuple_b))
```

```python
# tuple에서 특정 위치의 데이터 추출 (index값은 0부터 시작)
print(tuple_a[0])  # 1
print(tuple_b[1])  # 3.14
```

```python
# tuple 간의 합 -> 문자의 합과 같은 결과 => 새로운 튜플이 생성됨
tuple_a + tuple_b
```

```python
# tuple과 숫자의 합 => 불가능
# tuple_b + 1     # -> TypeError!
```

### 2. 리스트

```python
# list 형태의 데이터
list_a = [30, 20, 50, 10]
list_b = ['abc', 'fi', 'be']
```

```python
# list의 함수 중 정렬을 변경하는 함수
# sort() : 리스트의 원소들을 오름차순 정렬로 변경
# reverse() : 리스트의 원소들을 역순 정렬로 변경
# 위의 2개의 함수는 list를 변경시킴

list_a.sort()
list_a  # 오름차순 정렬로 바뀜
```

```python
# 내림차순 정렬은?
# 오름차순 정렬 후 역순으로 변경
# sort()는 매개변수 reverse 존재 (기본값이 False, 오름차순)
list_a.sort(reverse=True)      # reverse=True, 내림차순
list_a
```

```python
# 리스트에 원소를 추가하는 함수
# append('data') : 가장 마지막에 원소 추가
# insert(index, 'data') : 특정 위치에 데이터 추가

list_a.append(100)
print(list_a)
```

```python
list_a.insert(2,100)
print(list_a)
```

```python
# 리스트의 원소 수정

# 먼저 어떠한 데이터를 수정할지 선택해야 한다.
# list_a의 두 번째 위치의 데이터를 수정할 거라면, index = 1
print(list_a[1])  # 수정할 데이터 확인

# 30을 15로 변경하고 싶다면, 30의 데이터를 제거하고 15를 대입해야 한다.
list_a[1] = 15
print(list_a[1])
```

```python
# 리스트의 원소를 삭제하는 함수
# pop() : 가장 마지막 데이터 제거
# remove('data') : 특정 값의 데이터 제거
# del list[index] : 특정 위치의 데이터 제거

list_a.pop()
list_a
```

```python
list_a.append(15)
print(list_a)

# remove() 함수를 이용하여 list_a의 15라는 값을 제거
list_a.remove(15)
print(list_a)
```

```python
# 특정 위치의 데이터를 제거하는 del
del list_a[3]
list_a
```

```python
# 복사본 생성 (백업)
a = 10
b = a
print(a)
print(b)
```

```python
# 딕셔너리에 데이터 추가 -> key:value 형태로 추가
# 추가 기능과 수정 기능의 코드 차이가 거의 없다.
# 추가 -> 새로운 key = data
# 수정 -> 기존 key = data

# dict_a에 key 값은 loc, value 값은 seoul 데이터 추가
dict_a['loc'] = 'seoul'

dict_a
```

```python
# 기존의 key 값을 선택하고 데이터를 대입하면, 기존의 value는 제거되고 새로운 value가 추가됨
# = 데이터 수정
# dict_a의 key 값 'age'의 기존 value 값 20을 30으로 수정
dict_a['age'] = 30

dict_a
```

```python
# sort() -> 리스트의 각각의 원소들을 비교하여 가장 작은 데이터를 처음으로,
# 가장 큰 데이터를 마지막으로 하여 정렬을 변경하는 함수

sample_list = [30, 10, 50, 7]
# sample_list의 가장 작은 데이터를 담아줄 수 있는 새로운 빈 리스트 필요
sort_list = []

# sample_list의 길이(4)만큼 반복하는 반복문 생성
for i in range(len(sample_list)):
    # 첫 번째 원소의 데이터를 최솟값으로 지정
    min_value = sample_list[0]
    # 두 번째 원소부터 마지막 원소(len)까지 min_value와 비교해, 작은 데이터를 min_value에 대입
    for j in range(1, len(sample_list)):  # 3번 반복
        # min_value와 sample_list[j] 데이터를 비교해 작은 값 찾기
        if sample_list[j] <= min_value:
            min_value = sample_list[j]
    sample_list.remove(min_value)
    sort_list.append(min_value)
    print(sort_list)
    print(sample_list)       
    print(min_value)
```

```python

sample_list = [30, 10, 50, 7]
sort_list = []

for i in range(len(sample_list)):
    # 반복할 때마다 sample_list의 가장 작은 데이터 찾기
    min_value = min(sample_list)
    print(min_value)
    sort_list.append(min_value)
    # sample_list에서 min_value 데이터를 제거
    sample_list.remove(min_value)
print(sort_list)
```
