```python
# 인자의 개수가 가변인 경우
def func_7( *args ):    # * 붙이면 됨. 변수 이름은 상관없음
    # args의 타입?
    print(type(args))   # 함수 내에서 호출할 땐 * 붙일 필요 없음

    # 인자들의 누적 합 구해서 되돌려주는 기능
    # 합계 초기값 지정
    result = 0

    # args의 원소 개수만큼 반복 실행하는 반복문 생성
    for i in args:
        # i는 args의 원소들이 각각 대입됨
        result += i
    return result
```

```python
func_7(1, 2, 3, 4, 5)   # args는 튜플 형태

# 휘발성 데이터를 사용하는 이유가 적은 메모리 사용을 위한 것이 때문에
# 일반적으론 리스트보다 메모리 사용이 적은 튜플로 만들어짐.
```

```python
# 인자의 개수가 가변인 매개변수와, 개수가 1개인 매개변수가 존재하는 함수
def func_9(*args, data):   # 'args는 가변인 인자값, data는 1개의 인자값을 받겠다'는 뜻
    result = 0
    # data가 1이라면 누적합을 되돌려주고
    if data == 1:
        result = 0   # 초기값
        for i in args:
            result += i
    # data가 0이라면 누적곱을 되돌려준다.
    if data == 0:
        result = 1   # 초기값
        for i in args:
            result *= i
    return result
```

### 연습
- 함수를 생성
    - 매개변수 2개
        - _datas: 인자의 개수가 가변인 매개변수
        - -type: 기본값('max')이 설정된 매개변수
    - _type이 'max'라면 _datas들 중 가장 큰 값을 되돌려준다.
    - _type이 'min'이라면 _datas들 중 가장 작은 값을 되돌려준다.
    - _type이 'mean'이라면 _datas들의 평균을 되돌려준다.
    - 위의 3가지의 경우가 아니라면 "_type이 맞지 않습니다"를 되돌려준다.


```python
def func_11( *_datas, _type='max'):
    # _type의 데이터가 어떤 것인가?
    if _type == 'max' :   # 기본값
        # _datas들 중 가장 큰 값을 되돌려준다.
        return max(_datas)
    if _type == 'min' :
        return min(_datas)
    if _type == 'mean' :
        return sum(_datas) / len(_datas)
    return "type이 맞지 않습니다."   # if문이 모두 실행되지 않으면 return이 실행되는 구조

# input _변수
# input 변수
# _변수
# 지역변수 만드는 법은 다양함. 회사 가게 되면 그 회사에서 어떻게 사용하는지 꼭 인지해야 함.
```

```python
# 위와 결과는 같음. 그저 return을 적게 쓰는 방법.
def func_12(*_datas, _type='max'):
    if _type == 'max' :
        result = max(_datas)
    if _type == 'min' :
        result = min(_datas)
    if _type == 'mean' :
        result = sum(_datas) / len(_datas)
    else:
        result = "type이 맞지 않습니다."
    return result
```

```python
func_11(1,2,3,4,5)
```

```python
func_11(1,2,3,4,5, 'min')
# int와 str의 크기 비교는 불가능하다는 TypeError 발생
# 'min'이 첫 번째 매개변수에 들어갔기 때문
```

```python
func_11(1,2,3,4,5, _type='min')
# _type에 들어감을 명시해주면 됨
```

```python
func_11(1,2,3,4,5, _type='mean')
```

```python
func_11(1,2,3,4,5, _type='sum')
# 사용자 입장에서는 "type이 맞지 않습니다."보다
# 사용 가능한 타입을 알게 되면 좋음 (코드 수정, 리뷰 등등에서)
```
