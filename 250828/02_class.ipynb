### class 선언
    - class 키워드를 이용해서 class 선언
    - class의 이름은 일반적으로 단어의 첫글자를 대문자로 표기
    (소문자로 써도 문제가 되는 건 아니지만, 함수와 헷갈릴까봐 대문자로 많이 표기함.)

### class 생성
    - 선언이 된 class를 특정 공간에 저장하는 과정
    (변수를 생성해서 class를 대입)
    - 생성된 class 안에는 변수와 함수가 존재
    (같은 변수명을 사용해도 각각의 class 내에 저장됨)
### class 내부의 변수, 함수 호출
    - class가 저장되어 있는 공간(변수명) + ~안에(.) + class에 존재하는 변수(변수명), 함수(함수명()) 호출

```python
# 가장 기본적인 class 선언
class Class_1 :
    # 생성자 함수(초기화 함수)
    # class 내부에 함수를 생성할 때는 일반적으로 매개변수 self는 기본
    # self : 자기 자신(=생성되는 class 공간)
    # __init__ 함수 : class가 생성될 때 한 번 실행되는 함수로,
    #                따로 호출할 필요 X(매우 드뭄)
    def __init__(self, _a, _b):
        # class 내부에서 사용할 변수들을 저장하는 기능
        # self.a -> self(자기자신) + .(안에) + a(변수명)
            # 자기자신 안에 있는 a라는 공간  # 지역 안의 지역
        self.a = _a
        self.b = _b
        # __init__ 함수를 class 내부에서 사용할 데이터를 저장하는 기능
        # return을 따로 만들지는 않는다.

    def add_value(self):
        result = self.a + self.b
        return result
```

```python
# class 생성
# Class_1을 생성하면 __init__ 함수가 호출됨
# __init__ 함수의 매개변수 3개 -> 입력값도 3개를 넣어줘야 함
test = Class_1(10, 7)  # 입력값을 2개만 넣는다면? -> 실행됨
```

```python
test = Class_1(10, 7, 1)   # 입력값 3개 -> Error 발생

# 저장될 수 있는 건 3개인데, 4개가 입력되었다고 함.
# 기본값이 없는 매개변수라면 인자와 개수를 맞춰야 실행되는데,
# __init__의 매개변수 self, _a, _b 중 10은 _a, 7은 _b에 대응됨
# self에는 변수의 위치가 자동으로 입력되므로,
# 3번째 인자는 저장될 수 있는 공간이 존재하지 않아 Error 발생
# 인자는 2개를 입력해야 함
```

```python
test2 = Class_1(20, 4)

# 같은 class를 두 군데 저장한 것.
# 한 군데는 10, 7
# 다른 한 군데는 20, 4가 저장된 상태
```

```python
# class 내부에 있는 변수를 호출
# 생성이 된 class의 위치(변수명) + ~안에(.) + 변수명
print(test.a)   # test에 있는 a
print(test2.a)  # test2에 있는 a
```

```python
# class 내부에 있는 함수를 호출
# 생성이 된 class의 위치(변수명) + ~안에(.) + 함수()
# add_value() 함수를 호출
test.add_value()
```

```python
test2.add_value()
```

```python
# 계산기 class 선언
class Calculate:
    # 생성자 함수 -> 두 개의 숫자 데이터를 저장하기 위한 기능
    # 생성자 함수는 필수 X. 초기 데이터 저장이 필요 없으면 일반 함수 만들어서 저장해도 됨.
    def __init__(self, _a, _b):
        self.a = _a
        self.b = _b
    
    # 두 개의 데이터를 더한 뒤 되돌려주는 함수
    def add(self):
        result = self.a + self.b
        return result
    # 두 개의 데이터를 뺀 뒤 되돌려주는 함수
    def sub(self):
        result = self.a - self.b
        return result
    # 두 개의 데이터를 곱한 뒤 되돌려주는 함수
    def multi(self):
        result = self.a * self.b
        return result
    # 두 개의 데이터를 나눈 뒤 되돌려주는 함수
    def dev(self):
        result = self.a / self.b
        return result
    
    # result에는 self 붙이지 않는 이유
    # 클래스 내부에 만들던, 외부에 만들던 한 번 사용하고 사라지는 휘발성 변수이므로 귀찮아서.
```

```python
cal_1 = Calculate(20, 4)
cal_2 = Calculate(5, 10)
```

```python
print(cal_1.add(), cal_1.sub(), cal_1.multi(), cal_1.dev())
```

```python
print(cal_2.add(), cal_2.sub(), cal_2.multi(), cal_2.dev())

# 꿀팁
# Alt 누르는 위치마다 커서가 생성되어 한 번에 수정 가능
# 더블클릭해서 단어 선택 -> Alt 누르고 다중선택 -> 한 번에 수정 가능
```

```python
# cal_1, cal_2 함수들을 호출할 때 출력이 나오는 것은
# class 내부의 데이터가 변경이 되지 않는다는 것을 의미한다.
```

```python
# class 내부의 변수에 데이터를 변경하는 class 선언
class Class_2:
    def __init__(self, _a):
        self.a = _a
    # self.a의 연산(더하기)하는 함수 -> 매개변수 2개 추가
    def func_1(self, _val, inplace = False):  # 필수항목 1개(_val), 선택항목 2개
        # inplace가 True라면, self.a에 _val를 더하고 self.a를 변경
        if inplace:
            self.a += _val     # 출력이 안 나오게 하기 위해 return 없음
        # inplace가 False라면, self.a에 _val를 더한 데이터를 되돌려준다.
        else:
            result = self.a + _val
            return result
```

```python
test3 = Class_2(10)
```

```python
# test3 내부에 있는 func_1() 함수를 호출
# 해당 함수는 매개변수 3개 -> self (test3의 위치가 자동 입력됨)
#                          _val (필수 입력)
#                          inplace (기본값 False)
test3.func_1
```

```python
test3.a
```

```python
# func_1() 호출하는데 inplace를 True로 변경
test3.func_1(5, inplace=True)
```

```python
# class 변수
# 같은 class를 생성하는 경우 서로 공유하는 변수
class Class_3:
    # self 변수들의 생성위치 -> 함수 안에서 생성
    # class 변수들의 생성위치 -> 함수 밖에서 생성
    cnt = 0      # class 변수1
    users = []   # class 변수2   총 2개
    # 유저를 등록하는 함수
    def add_user(self, _name):
        # 유저가 등록됨에 따라 cnt 1씩 증가
        self.cnt += 1
        # 유저의 이름 users에 추가
        self.users.append(_name)  # users.append(_name)만 입력하면 users라는 변수를 찾을 수 없다고 함.
                                  # 따라서 cnt와 users에 self를 붙여봄
```

```python
# 생성자 함수가 없으니 입력값도 필요 없음
user1 = Class_3()
user2 = Class_3()
```

```python
# user1에 add_user() 함수 호출
user1.add_user('kim')
```

```python
user2.add_user('park')
```

```python
print(user1.cnt, user1.users)
print(user2.cnt, user2.users)

# 원래 안 되는 게 맞음
# self는 만들어진 공간(지역) 내에서 사용하고, class는 그 상위공간.
# 따라서 class에서 불러와야 함.
```

```python
cnt = 0       # 단일 데이터
users = []    # 자료형 데이터

# 자료형 데이터 = 자료형 데이터 -> 이름이 하나 더 붙는 형태이므로
# 복제해서 사용해줘야 함

user1_cnt = cnt
user1_users = users

user1_cnt += 1
user1_users.append()  # kim

user2_cnt = cnt
user2_users = users

user2_cnt += 1
user2_users.append()  # park
```

```python
# class 변수
# 같은 class를 생성하는 경우 서로 공유하는 변수
class Class_3:
    # self 변수들의 생성위치 -> 함수 안에서 생성
    # class 변수들의 생성위치 -> 함수 밖에서 생성
    cnt = 0      # class 변수1
    users = []   # class 변수2   총 2개
    # 유저를 등록하는 함수
    def add_user(self, _name):
        # 유저가 등록됨에 따라 cnt 1씩 증가
        Class_3.cnt += 1
        # 유저의 이름 users에 추가
        Class_3.users.append(_name)
        ### self가 아닌 Class에서 불러와야 공유됨 ###
```
