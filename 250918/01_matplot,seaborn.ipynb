```python
# legend(): matplot에 범례 추가
# grid(): 보조선 추가
# title(): 그래프 제목 추가
# xlabel(), ylabel(): x축, y축 이름 지정
x = np.linspace(0, 10, 200)   # 0부터 10까지 200개의 데이터 생성
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'Sin')
plt.plot(x, y2, label = 'Cos')

# 그래프에 보조선 격자 추가
# grid( 격자표시여부, 보조선 축 )
plt.grid( True, axis='both' )

# x축, y축 이름 생성
plt.xlabel('X')
plt.ylabel('Y')

# 그래프 제목
plt.title('sin, cos graph')

# 범례 추가
# plt.legend(['Sin', 'Cos'])   # plt.plot(x, y1) 처럼 그래프에 설명이 따로 없을 때 부가적으로 추가하는 경우
# plt.legend()                 # plt.plot(x, y1, label = 'Sin') 처럼 그래프에 설명이 붙어있을 때 라벨을 그대로 가져와 추가하는 경우 (이런 경우가 더 많음!)
# legend()에서 범례의 위치를 변경하는 매개변수는 loc (기본값은 best, 즉 그래프를 가리지 않는 위치에 자동 배치)
#   사용 예시: subplot()으로 여러 개의 그래프를 그릴 때, loc를 지정하지 않으면 범례 위치가 제각각
# plt.legend(loc=5)  # 숫자도 가능. 문자도 가능.
plt.legend(loc = 'upper right')

plt.show()
```

```python
# rc: 환경설정
if platform.system() == 'Darwin':
    plt.rc('font', family= 'AppleGothic')
else:
    plt.rc('font', family= 'Malgun Gothic')
```

```python
# 막대 그래프
x = ['국어', '영어', '수학', '과학']
scores = [80, 60, 70, 50]

plt.bar(x, scores)

# x축의 데이터들을 회전 -> xticks(): x축 데이터들에 접근하는 함수
# rotation: 데이터명 회전  (보통 이름이 길면 45도 회전)
plt.xticks(rotation= -45)

# 그래프에 문자 추가 (주석과는 다른 함수 사용)
# text( x축, y축, 문자 )
for i in range(len(scores)):
    plt.text( i, scores[i]+1 , scores[i] )  # scores[i]+1: 막대랑 너무 붙어있어서 조금 올려줌
```

```python
# 히스토그램
datas = np.random.random(1000)

# 히스토그램의 구간은 기본적으로 10개로 생성
# bins: 구간 개수 변경 가능
plt.hist(datas, bins= 15)

plt.show()
```

```python
# 산점도
x = np.random.random(1000) # 0부터 1까지 1000개의 랜덤한 모수
y = np.random.random(1000) # 0부터 1까지 1000개의 랜덤한 모수
s_data = np.random.random(1000) * 100  # 0부터 100까지 1000개의 랜덤한 모수

# 그래프 사이즈 변경
plt.figure(figsize= (10, 8))


# 값들이 중복되어 점이 겹쳐져 있는지 확인해봐야 한다.
# alpha: 투명도  (굉장히 많이 사용됨)
# s: scale의 약자(점 크기 변경)
# c: color의 약자(색깔 변경)
plt.scatter(x, y, alpha= 0.5, s= s_data, c= 'green')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

# seaborn을 통해 더 편하게 plot을 그릴 수 있다.
# hue: 종류별로 나눠서 보여준다.
sns.scatterplot(data= iris_data,
                x= 'sepal_length', y= 'sepal_width',
                hue= 'species'  # 범례
                )

plt.show()
```

```python
# 박스 플롯
plt.figure(figsize= (12, 6))

sns.boxplot( data= iris_data, x= 'species', y= 'petal_width')

plt.show()
```

```python
# pearson 상관계수 = 가장 기본적. spears 상관계수 = 종종 사용.

drinks = pd. read_csv('../csv/drinks.csv')
drinks.info()
```

```python
corr = drinks.loc[:, 'beer_servings':'total_litres_of_pure_alcohol'].corr()
corr
```

```python
plt.figure(figsize= (8, 5))

sns.heatmap(
    data= corr,
    annot= True,             # annotate(주석) 유무. 안에 상관계수를 표시해준다. (기본값 False)
    square= True,            # 히트맵 정사각형으로 표시 여부 (기본값 False)
    fmt= '0.3f',             # 주석의 소수점 자리수 표시
    annot_kws= {'size': 20}, # 주석의 환경설정. 폰트 크기 지정.
    cbar= False,             # 색상 바 표시 여부 (기본값 True)
    xticklabels= ['beer', 'sprit', 'wine', 'alcohol'],  # x축 이름 변경
    yticklabels= ['beer', 'sprit', 'wine', 'alcohol']   # y축 이름 변경
)

plt.show()
```

```python
sns.pairplot(
    data= drinks.loc[:, 'beer_servings':'total_litres_of_pure_alcohol'],
    kind= 'reg'    # 추세선 추가
)

plt.show()
```

```python
# subplots(): 그래프가 그려지는 공간을 나눠서 여러 개의 그래프를 그려주는 함수
# plt.subplot(2, 2, 1)
fig, axes = plt.subplots(2, 2)      # fig, axs : 일반적으로 사용하는 변수명

# axes[0][0].plot([1,2,3,4])
# axes[0][1].plot([4,3,2,1])     # 귀찮음!

axes.flat[3].plot( [1,2,3,4] )

plt.show()
```

```python
# beer_servings와 나머지 3개의 컬럼들 간 상관관계
cols = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
fig, axes = plt.subplots( 3, 1, figsize= (5, 15) )
for i in range(1, len(cols)):
    ax = axes.flat[i-1]
    ax.scatter( drinks[cols[0]], drinks[cols[i]] )

plt.show()
```
