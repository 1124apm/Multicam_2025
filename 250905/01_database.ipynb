- MySQL과 Python 연동
    1. 연동 기능이 존재하는 라이브러리(pymysql) 설치
    2. 라이브러리 로드
    3. 데이터베이스 서버와의 연결
        - 서버의 정보 필요
            - 주소
            - port
            - 사용자 아이디
            - 비밀번호
            - 사용할 데이터베이스명(스키마)
    4. 커서 생성 
        - 커서: 데이터를 저장, 수정하는 가상의 (테스트)공간
            - 원본의 데이터를 안정적으로 유지하는 데 용이함
    5. query문을 작성하여 질의를 한다.
        - insert, update, delete: return이 존재하지 않는다.
        - select: return이 존재한다.
            - execute(  ): cursor로 보낸다.
            - fetchall(  ): cursor에서 Python에 데이터를 가져온 뒤, cursor에서는 삭제한다.
    6. (작업완료 후) DB server와 Cursor 동기화
    7. DB server와의 연결 종료

```python
# 라이브러리 로드
import pymysql
```

```python
# DB server(_db)와의 연결
# = 연결되는 정보를 변수에 저장한다.

_db = pymysql.connect(
    # host= 'localhost' # (내 컴퓨터)
    host= '127.0.0.1',  # 서버의 주소
    port= 3306,         # str로 쓰지 않게 주의
    user= 'root',       # 유저의 아이디
    password= '1408',   # 비밀번호 (비밀번호가 없을 시 '')
    db = 'multicam'
)
```

```python
# RuntimeError: 'cryptography' package is required for sha256_password or caching_sha2_password auth methods
# 위와 같은 cryptography 에러 발생 시
# !pip install cryptography

# 패스워드 암호화해주는 라이브러리
```

```python
# connect는 결과로 클래스를 생성하는 구조
# _db를 이용하여 Cursor 생성
# case 1. 기본값으로 커서 생성 (일반적인 샘플 코드에 쓰이는 방식, 비추천)
    # select문을 이용해 데이터를 로드할 때, 데이터의 타입이 tuple -> 2차원 데이터를 튜플로 받으면, () 안의 () -> column name이 들어오려면 3차원이어야 한다.
cursor = _db.cursor()

# case 2. DictCursor 옵션을 사용해 커서 생성 (추천)
    # select문을 이용해 데이터를 로드할 때, 데이터의 타입이 dict -> column name을 key에 받으면 된다.
cursor2 = _db.cursor( pymysql.cursors.DictCursor )
```

```python
# query문을 작성해 커서에 질문을 보낸다.
select_query = """
    select
    *
    from
    `emp`     
"""
# `emp`의 ` 생략 가능하지만 '로 쓰면 Error
# select_query를 커서에게 보낸다. (Cursor.execute() 함수 사용)
cursor.execute(select_query)
cursor2.execute(select_query)
# select에 대한 index 개수가 출력됨
```

```python
cursor.fetchall()  # 2차원 형태의 튜플
```

```python
cursor2.fetchall()  # 2차원 리스트 안의 딕셔너리 형태(key값이 column의 이름)
```

```python
insert_query = """
    insert into
    `user_info`
    values
    ("test2", "1234", "park", "000101")    
"""
# 커서에게 insert_query 질의를 보낸다.
cursor2.execute(insert_query)
```

```python
select_query2 = """
    select * from `user_info`
"""
cursor2.execute(select_query2)
cursor2.fetchall()
```

```python
# 현재까지의 작업은 커서(가상공간)에 데이터의 변화만 존재
# cursor와 DBserver와의 동기화 필요
_db.commit()
```

```python
# 데이터베이스와의 연결을 종료
_db.close()
# close 안 하면? = 식당에서 밥을 다 먹고 그 자리에서 아무것도 안 하는 것

# 연결(connect)하고 안 닫으면(close) 안 끊김
# 연결(request)하고 응답(reponse)받으면 자동으로 끊김
```
