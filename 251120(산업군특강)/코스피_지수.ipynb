```python
import pandas as pd
import numpy as np
```

```python
kospi=pd.read_csv('코스피_지수.csv', encoding='cp949')
```

```python
X= kospi.drop(['코스피지수'], axis=1)
```

```python
X.head()
```

```python
y=kospi['코스피지수']
```

```python
y.tail()
```

# 훈련 테스트 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
X_train.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

```python
X_train.head()
```

```python
columns = X_train.columns
columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
X_train=std.fit_transform(X_train)
X_test=std.transform(X_test)  # 여기서는 fit하면 안됨.
```

# 1. 선형회귀모델

```python
# 1. 인스턴스화
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 2. 적합화
lr.fit(X_train, y_train)
```

```python
# 3. 예측
y_pred = lr.predict(X_test)
```

```python
# 4. 평가
from sklearn.metrics import mean_squared_error as mse, r2_score as r2
import math
```

```python
# RMSE와 R^2 구하라
print('RMSE: ', math.sqrt(mse(y_pred, y_test)))
print('R^2: ', r2(y_pred, y_test)*100)
```

```python
np.mean(y_test)
```

```python
107/1500
```

```python
# 2_1. 계수 프린트
c1=lr.coef_.reshape(1,-1)
c2=pd.DataFrame(c1, columns=columns)
c2.T
```

# 2. 릿지 회귀

```python
#1. 인스턴스화
from sklearn.linear_model import Ridge
ridge_model= Ridge(alpha=100)

# 2. 적합화
ridge_model.fit(X_train, y_train)

# 3. 예측
y_pred = ridge_model.predict(X_test)

# 4. 평가 RMSE와 R^2
from sklearn.metrics import mean_squared_error as mse, r2_score as r2
import math
print(math.sqrt(mse(y_pred, y_test))) # RMSE
print(r2(y_pred, y_test)*100) #R^2

# 2_1. 계수 프린트
c1=ridge_model.coef_.reshape(1,-1)
c2=pd.DataFrame(c1, columns=columns)
c2.T
```

```python
c1.shape
```

# 3. 라쏘 회귀

```python
#1. 인스턴스화
from sklearn.linear_model import Lasso
lasso_model= Lasso(alpha=100)

# 2. 적합화(학습)
lasso_model.fit(X_train, y_train)

# 3. 예측
y_pred = lasso_model.predict(X_test)

# 4. 평가 RMSE와 R^2
from sklearn.metrics import mean_squared_error as mse, r2_score as r2
import math
print(math.sqrt(mse(y_pred, y_test))) # RMSE
print(r2(y_pred, y_test)*100) #R^2

# 2_1. 계수 프린트
c1=lasso_model.coef_.reshape(1,-1)
c2=pd.DataFrame(c1, columns=columns)
c2.T
```

# 다항식 회귀모델

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 인스턴스화
poly= PolynomialFeatures(degree=2)
```

```python
# X를 변환한다.
poly_X_train=poly.fit_transform(X_train)  # fit_transform => fit + transfomr 같이 사용한 것임.
```

```python
poly_X_test=poly.transform(X_test)
```

```python
poly_X_train.shape
```

```python
lr=LinearRegression()
lr.fit(poly_X_train, y_train)
y_pred=lr.predict(poly_X_test)
print(math.sqrt(mse(y_pred, y_test)))
print(r2(y_pred, y_test))
```
