# 다항회귀
- 데이터가 단순히 직선의 형태가 아닌 비선형의 형태를 가지고 있는 경우,</br> 각 변수의 거듭제곱을 새로운 변수로 추가하면 선형 모델로 사용 가능
- 이렇게 확장된 특성을 데이터셋에 선형 모델로 훈련시키는 방법

```python
import pandas as pd
```

```python
cereal = pd.read_csv("../data/cereal.csv")
```

```python
cereal.info()
```

```python
cereal.head()
```

```python
# 문자형 데이터를 제외
cereal = cereal.iloc[:, 3:]
cereal.head(1)
```

```python
# 설탕의 함유량에 따른 영양등급 평가에 대한 회귀분석
cereal.describe()
```

```python
# sugars 컬럼의 최솟값이 음수 -> 음수인 데이터 개수 확인
# sugars의 데이터가 음수인 조건식 생성
(cereal['sugars'] < 0).sum()
    # np.int64(1)
flag = cereal['sugars'] < 0

cereal.loc[flag, ]
# 인덱스 57번의 데이터가 출력됨
```

```python
# drop() 함수를 이용하여 특정 인덱스 제거
cereal.drop(57, axis=0)
```

```python
# 음수 조건식에 부정 (비트 연산자를 이용하여 부정 처리)
cereal = cereal.loc[~flag, ]
```

```python
# scatter 그래프를 이용하여 분포 확인
import matplotlib.pyplot as plt

x = cereal['sugars']
y = cereal['rating']

plt.scatter(x, y)
```

```python
# 산점도 그래프에 회귀선 추가
import seaborn as sns

sns.regplot(data= cereal, x= 'sugars', y= 'rating', order=2)

plt.show()
```

```python
# sugars, rating 두 개의 컬럼을 이용하여 선형회귀 모델 학습
cereal2 = cereal[ ['sugars', 'rating'] ]
```

```python
x = cereal2['sugars'].values
y = cereal2['rating'].values
```

```python
# 행렬 구조를 변경
    # 예) 데이터 76개, x.reshape(38, 2) --> 38행 2열
# x.reshape(-1, 1) --> 행의 수는 그대로 두고 하나씩 2차원 행렬 데이터로 만들기
x.reshape(-1, 1)
```

```python
# train, test로 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size= 0.3, random_state= 42
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

lr = LinearRegression()

lr.fit(X_train.reshape(-1,1), Y_train)

pred_1 = lr.predict(X_test.reshape(-1,1))

mae_1 = mean_absolute_error(Y_test, pred_1)

print(round(mae_1, 4))
```

---

```python
# n차원 다항 조건 생성
from sklearn.preprocessing import PolynomialFeatures
```

범위 지정은 train
trasform은 train과 test 모두 같이 작업
데이터에 누수 생길 수 있으니 test에는 따로 핏하지 않음

```python
# PolynomialFeatures class 생성
# degree: 차수 지정
poly_reg = PolynomialFeatures(degree= 2)

# poly_reg의 범위를 지정하고 데이터를 변환하여 되돌려준다.
X_train_poly = poly_reg.fit_transform(X_train.reshape(-1, 1))
X_test_poly = poly_reg.transform(X_test.reshape(-1, 1))

# 딘슨 선형 회귀 모델 생성
lr = LinearRegression()
# 변환한 값으로 학습시켰으니, 변환한 값으로 예측해야 한다.
lr.fit(X_train_poly, Y_train)
pred_2 = lr.predict(X_test_poly)

mae_2 = mean_absolute_error(Y_test, pred_2)
```

```python
print( round(mae_2, 4) )
```
