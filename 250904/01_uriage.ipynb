< 대소문자 문제 - 모든 문자를 대문자로 변경 > </br>
case 2: map(  ) 함수 사용

```python
df2 = uriage.copy()
```

```python
# Series.map() 함수를 이용해 데이터 변환
# map() -> 1차원 데이터에서 각각의 원소들을 함수에 대입한 결과로 새로운 1차원 데이터 생성
# map에서는 일반 def 키워드 함수(이름 존재, 재사용), lambda 함수(휘발성) 사용 가능
# 이 경우 간단한 함수이므로 lambda 함수 사용이 더 효율적

# lambda 함수 사용 시
df2['item_name'].map(
    # x의 타입은 str -> 문자열 내장함수인 upper() 사용 가능
    lambda x: x.upper()
)
```

```python
# map() + def 함수 사용 시
def change(x):
    # x에 대입되는 데이터는 타입이 str인 데이터 -> 문자열 내장함수 사용 가능
    x.upper()

df2['item_name'].map(
    change   # ()를 넣으면 함수를 호출하는 것이므로, 함수명만 적어야 한다.
)

# change 함수에 return이 없어, 모든 값이 None으로 출력된다.
```

```python
# 따라서 return이 꼭 필요하다.

# map() + def 함수 사용 시
def change(x):
    # x에 대입되는 데이터는 타입이 str인 데이터 -> 문자열 내장함수 사용 가능
    x = x.upper()
    return x

# df2가 아닌 df2['item_name']에 넣어야 함을 주의
# 아니면 작업 끝내고 df2['item_name'] Series 한 개만 남음 ㅠㅠ
df2['item_name'] = df2['item_name'].map(
    change   # ()를 넣으면 함수를 호출하는 것이므로, 함수명만 적어야 한다.
)
```

< item_price 데이터에 결측치 채우기 >
- 상품별 결측치를 제외한 item_price의 평균을 구한다.
- 해당 상품의 item_price가 결측치인 데이터에 평균값을 대입한다.    

```python
# 상품A의 item_price가 결측치가 아닌 데이터를 확인한다.
# 조건식: item_name이 상품A ( uriage['item_name'] == 상품 A )
#        item_price가 결측치 ( uriage['item_price'].isna() ) --> 이 조건을 부정

flag = uriage['item_name'] == '상품A'
flag_null = uriage['item_price'].isna()
```

```python
# item_name이 상품A(flag)이고(and) item_price가 결측치가 아닌(~flag_null)
# 상품 정보(모든 컬럼) 확인 -> 컬럼의 조건식은 필요X
uriage.loc[
    flag & ~flag_null,
]
# 상품A의 정보만 나오고, item_price도 결측치가 아님

# 위의 인덱스 조건에서 item_price 컬럼(컬럼의 조건식)의 데이터의 평균값(mean())을 구한다.
mean_price = uriage.loc[
    flag & ~flag_null,
    'item_price'
].mean()
```

```python
# mean_price를 item_name이 상품A(flag) 이고, item_price가 결측치(flag_null)인
# item_price에 (컬럼의 조건) 대입
uriage.loc[ flag & flag_null, 'item_price'].fillna(mean_price, inplace=True)
# 결측치만 있는 Series 데이터의 결측치를 평균값으로 채운다.
```

```python
# inplace=True 했음에도 원본 데이터 수정되지 않은 채 그대로 NaN
# fillna 쓸 때 처음부터 끝까지는 가능한데, 중간만 수정하는 건 inplace가 작동X
uriage.loc[ flag & flag_null, 'item_price'] \
     = uriage.loc[ flag & flag_null, 'item_price'].fillna(mean_price)
```

```python
uriage.loc[ flag & flag_null, 'item_price']
# 모두 100으로 잘 바뀌어 있다.
```

```python
# 위의 작업들을 상품A부터 상품Z까지(item_name 데이터에서 중복으로 제거한 리스트) 반복 실행
item_list = uriage['item_name'].unique()

# 반복문 작성
# item_list의 길이만큼 반복
for item_name in item_list:
# (다른 방법) for i in range(len(item_list)):
    # item_name 변수에는 item_list의 각각의 원소들이 대입
    # (다른 방법) i 변수에는 item_list의 index값이 대입
    flag = uriage['item_name'] == item_name  # ex. '상품A'
    # (다른 방법) flag = uriage['item_name'] == item_list[i]
    flag_null = uriage['item_price'].isna()

    mean_price = uriage.loc[flag & ~flag_null, 'item_price'].mean()
    # 또는
    uriage.loc[flag & flag_null, 'item_price'] = mean_price
```
