```python
# def func_6( *_numbers, _type ) 로 가변을 먼저 넣으면 ↓↓↓
def func_6( _type, *_numbers ):
    if _type == 'sum':
        result = 0
        for num in _numbers:
            result += num
    else:
        result = 1
        for num in _numbers:
            result *= num
    return result
```

### 특정 경로의 파일을 모두 로드하는 함수 생성
- 매개 변수 3개
    - file_path : 파일 경로 (str)
    - ext : 확장자 (str / 기본값 .csv)
    - concat : 결합 방식 (bool --> True면 결합, False면 결합 X / 기본값 True)
- file_path를 이용해 파일 목록 생성
    - file_path에 '/' 더해준다.
- 확장자에 따라서 새로운 목록 생성
- concat이
    - True라면 빈 데이터프레임 생성
    - False라면 빈 리스트 생성
- 새로운 목록을 이용하여 반복문 생성
    - 확장자에 따라서 read 함수를 지정하고 파일 로드
    - concat에 따라 concat()를 사용할지, append()를 사용할지 지정
- 결과물을 되돌려준다.

# 파일 분할 관리는 매우 중요!!!!!!!!!!!!!!!
- 결과를 뽑는 게 제일 중요하니, 하나를 일단 다 만들어둔다.
- 특정한 기능을 뽑아 따로 돌려보고 문제 없는지 확인한다.
- 계속 하다보면 이 부분은 분할해둬야겠다 하는 느낌이 생긴다.

```python

def data_load_os( file_path, ext= '.csv', concat= False):
    # file_path에 '/' 더해준다.
    file_path = file_path + '/'
    # file_path의 파일 목록 로드
    file_list = os.listdir(file_path)
    # ext에 맞는 확장자 파일만 존재하는 새로운 리스트 생성
    new_list = []
    for name in file_list:
        if name.endswith(ext):
            new_list.append(name)
    
    # concat에 따라서 빈 데이터프레임을 생성할지, 빈 리스트를 생성할지
    if concat:
        result = pd.DataFrame()
    else:
        result = []
    
    # 새로운 리스트로 반복 실행
    for name in new_list:
        # ext 확장자에 따라 read 함수 정하기
        if ext == '.csv':
            try:
                df = pd.read_csv(file_path + name)
            except:
                df = pd.read_(file_path + name, encoding='cp949')
        elif ext == '.json':
            try:
                df = pd.read_json(file_path + name)
            except:
                df = pd.read_(file_path + name, encoding='cp949')
        elif ext == '.xml':
            try:
                df = pd.read_xml(file_path + name)
            except:
                df = pd.read_(file_path + name, encoding='cp949')
        elif ext in ['.xlsx', '.xls']:
            try:
                df = pd.read_excel(file_path + name)
            except:
                df = pd.read_(file_path + name, encoding='cp949')
        else:
            return "사용 가능 확장자가 아닙니다."

        if concat:
            result = pd.concat( [result, df], axis= 0 )
        else:
            result.append( df )
    # 결과를 되돌려준다.
    return result
```

- 파일의 경로를 이용하여 파일 전체 목록 로드 (glob 사용)
- 로드한 데이터들이 저장될 빈 리스트 생성
- 목록을 이용하여 반복문 실행
- 파일의 확장자를 따로 추출 -> 확장자에 따라 read 함수를 지정
- 예외처리로 encoding 변환
- 로드가 된 데이터프레임을 빈 리스트에 추가
- 결과를 되돌려준다.

```python
def data_load_glob(file_path):
    file_list = glob(file_path + '/*.*')
    # 결과값인 빈 리스트 생성
    result = []
    for name in file_list:
        file_ext = os.path.splitext(name)[-1]
        if file_ext == '.csv':
            try:
                df = pd.read_csv(name)
            except:
                df = pd.read_csv(name, encoding='cp949')
        elif file_ext == '.json':
            try:
                df = pd.read_json(name)
            except:
                df = pd.read_json(name, encoding='cp949')
        elif file_ext in ['.xlsx', 'xls']:
            df = pd.read_excel(name)
        elif file_ext == '.xml':
            try:
                df = pd.read_xml(name)
            except:
                df = pd.read_xml(name, encoding='cp949')
        result.append(df)
    return result
```

```python
# glob에서 * 와 *.* 의 차이
glob('../csv/*')
glob('../csv/*.*')
# 데이터의 개수가 10개 가량 차이 난다.
# * : 폴더를 포함한 리스트 생성
# *.* : 폴더를 제외한, 확장자를 가진 파일만으로 리스트 생성
```
