#### 연습
- 함수(func_2) 생성
    - 매개변수 1개 -> n (기본값 존재 X)
    - n 팩토리얼(n!)의 값을 되돌려주는 함수

```python
# 함수 생성
# 매개변수에 함수의 결과에 대한 힌트를 포함
def func_2( n : int) -> int :
    # 팩토리얼: 1부터 시작해 자기 자신까지 1씩 증가시키며 누적곱
    result = 1
    for i in range(2, n+1):
        result *= i
    return result
```

```python
func_2(5)
```

```python
def func_3( n : int) -> int :
    # 팩토리얼: 1부터 시작해 자기 자신까지 1씩 증가시키며 누적곱
    result = n
    for i in range(n-1, 1, -1):
        result *= i
    return result
```

### 재귀 함수
- 어떤 이벤트에서 자기 자신을 포함하고 다시 자기 자신을 사용하여 정의되는 것을 의미한다. </br>
    = 내가 나를 호출해, 반복 실행할 수 있도록 함.
- 재귀를 효과적으로 사용하면, 프로그램을 간결하고 효율적으로 작성할 수 있다.

```python
# 팩토리얼 n!
def func_4(n: int) -> int:
    if n > 0:
        return n * func_4( n-1 )
    else:
        return 1
```

```python
# list 역순으로 뒤집기
test_list = [1,2,3,4,5]
```

```python
# list에 내장된 reverse() 함수를 생성
def list_reverse(_x):
    # 가장 마지막의 데이터를 추출하여 새로운 리스트에 추가
    result = []

    for i in range(1, len(_x)+1):        # 1부터 시작
    # for i in range(len(_x)-1, -1, -1):    # 또는 길이-1부터 시작
        result.append(
            _x[-i]
        )
    return result
```

```python
# list에서 백업 -> copy() 함수 이용
# _x = test_list 중 test_list는 함수가 호출될 때마다 변경되므로, _x를 new_list에 백업해야 한다.
# 따라서 ↓
def list_reverse2(_x):
    # _x의 복사본 생성
    result = _x.copy()
    for i in range(len(result)//2):
        result[i], result[-1-i] = result[-1-i], result[i]
    return result
```

### 연습
- 문자의 위치를 찾는 함수를 생성해보자.
    - 함수의 이름은 c_find()
    - 매개변수 2개: _data, _word
    - _data에서 _word의 위치를 출력하는 기능
    - 찾는 _word가 여러 개라면, 여러 개를 동시에 출력
    - 찾는 _word가 존재하지 않는다면, 비어있는 튜플 () 출력
- 테스트
    - c_find('apple', 'p')
    - c_find('apple', 'b')
    - c_find(['a','b','c','d'], 'a')

```python
# find(), index() 함수를 모른다는 기준으로 함수 구성

def c_find(_data, _word):
    # index들이 저장될 공간을 설정
    result1 = []   # 리스트 쓰는 경우
    result2 = ()  # 튜플 쓰는 경우
    for i in range(len(_data)):
        if _data[i] == _word:
            # result: list에 데이터 추가
            result1.append(i)

            # result2: 튜플에 데이터 추가 -> 불가 -> 새로 만들 수 밖에 없음
            # tuple + tuple => 추가의 형태로 새로운 tuple 생성됨
            # result2 += tuple([i])   <- 튜플 클래스 이용하거나, ↓
            result2 += (i,)
    return result1, result2
```

#### 연습
- 소수를 구하는 함수를 생성하시오
    - 매개변수 2개: _start (기본값 2), _end (기본값 100)
    - 소수란?
        - 1과 자기 자신 이외엔 나누어 떨어지지 않는 수
        - 조건: 2부터 시작해 (자기 자신 - 1)까지 어떤 정수로도 나누어 떨어지지 않는다.
    - _start부터 _end까지의 수를 반복해, 소수인 데이터를 비어있는 리스트에 추가한다.
    - 소수의 리스트와 반복 횟수를 반환한다.

```python
# 방법 1
# if-else 안 쓰는 경우 -> 진짜 반복횟수가 나옴 -> 횟수 훨씬 많음 -> 비효율적

# 함수 생성 (매개변수 2개 기본값 설정)
def prime_list(_start = 2, _end = 100):
    # 반복 횟수(초기값 0)
    cnt = 0
    # 비어있는 리스트 생성
    result = []
    # 반복문을 이용하여 _start부터 (_end+1) 까지 반복 
    for n in range(_start, _end+1):
        flag = 1
        # 2부터 (자기 자신-1)까지 반복
        for i in range(2, n):
            # 반복 횟수 1씩 증가
            cnt += 1
            # 나눈 나머지가 0인지 확인
            if n % i == 0:
                flag = 0
        # 조건문이 거짓이라면 
        if flag:  # 0을 제외한 숫자는 모두 True이므로 flag == 1 -> flag로 생략 가능
            # 소수 리스트에 추가
            result.append(n)
    return cnt, result
```

```python
# 방법 2
# 함수 생성 (매개변수 2개 기본값 설정)
def prime_list(_start = 2, _end = 100):
    # 반복 횟수(초기값 0)
    cnt = 0
    # 비어있는 리스트 생성
    result = []
    # 반복문을 이용하여 _start부터 (_end+1) 까지 반복 
    for n in range(_start, _end+1):
        # 2부터 (자기 자신-1)까지 반복
        for i in range(2, n):
            # 반복 횟수 1씩 증가
            cnt += 1
            # 나눈 나머지가 0인지 확인
            if n % i == 0:
                # 참이라면 반복문 종료
                break
        # 조건문이 거짓이라면 
        else :
            # 소수 리스트에 추가
            result.append(n)
    return cnt, result

# 어차피 둘 다 같은 반복문 안이므로 if-else 위치가 달라도 연결된다는 점 주의.
```

```python
# 방법 3
# 함수 생성 (매개변수 2개 기본값 설정)
def prime_list(_start = 2, _end = 100):
    # 반복 횟수(초기값 0)
    cnt = 0
    # 비어있는 리스트 생성 <- 2는 미리 포함
    result = [2]
    # 반복문을 이용하여 _start부터 (_end+1) 까지 반복
    # 짝수는 2를 제외한 나머지는 소수가 될 수 없으므로, 홀수만 반복 실행
    # _start가 짝수라면 _start + 1
    if _start % 2 == 0:
        _start += 1
    for n in range(_start, _end+1):
        # 2부터 (자기 자신-1)까지 반복
        for i in range(2, n):
            # 반복 횟수 1씩 증가
            cnt += 1
            # 나눈 나머지가 0인지 확인
            if n % i == 0:
                # 참이라면 반복문 종료
                break
        # 조건문이 거짓이라면 
        else :
            # 소수 리스트에 추가
            result.append(n)
    return cnt, result
```
