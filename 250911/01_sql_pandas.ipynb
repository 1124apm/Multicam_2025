### 데이터프레임의 결합
- 유니언 결합
    - 단순하게 행을 결합하거나 열을 결합하는 방식
    - pandas에 내장된 concat() 함수 이용
        - concat( [데이터프레임1, 데이터프레임2, ...] )
            - axis 매개변수
                - 0(index, rows) : 단순 행 결합
                - 1(columns) : 단순 열 결합
            - ignore_index 매개변수
                - False : 결합되는 기준(인덱스 or 컬럼)을 보존
                - True : 결합되는 기준(인덱스 or 컬럼)을 초기화
- 조인 결합
    - 특정한 조건에 맞춰서 열 결합
    - 데이터프레임 간의 결합 - 특정 컬럼의 데이터들이 같은 값들로 이루어져 있는 경우에 열을 추가하는 결합 방식
    - pandas, DataFrame에 내장된 merge() 함수 이용
    - 조건에 맞춰서 열을 결합하는 방식이므로, concat()과는 다르게 2개의 데이터프레임만 결합 가능</br>-> 더 많은 데이터프레임을 결합하려면 merge() 함수를 여러 번 사용하면 됨
        - pd.merge( 데이터프레임명, 데이터프레임명 )
        - 또는 df.merge( 데이터프레임명 )
            - on 매개변수
                - 조건식(두 개의 데이터프레임이 공통적으로 가지고 있는 데이터 컬럼명)
                - 두 개의 데이터프레임이 공통적으로 가지고 있으나 컬럼명이 다른 경우?
                    - left_on 매개변수: 첫 번째 데이터프레임의 컬럼명
                    - right_on 매개변수: 두 번째 데이터프레임의 컬럼명
            - how 매개변수
                - 조인 결합 방식
                    - inner: 두 개의 데이터프레임이 공통적으로 가지고 있는 데이터만 결합하여 출력
                    - outer: 두 개의 데이터프레임의 합집합
                    - left: 왼쪽(첫 번째) 데이터프레임을 기준으로 결합
                    - right: 오른쪽(두 번째) 데이터프레임을 기준으로 결합

```python
# query_11 = """
#     select * from
#     `tran_1` left join `tran_d_1`
#     on 'tran_1`.`transaction_id` = `tran_d_1`.`transaction_id`
# """

# 별칭으로 쓰고 싶으면 'as 별칭' 사용 가능 (생략도 가능)
# `t1`. 도 생략 가능
# 이름 바꾸기도 가능 (`거래 식별자`)
query_11 = """
    select
    `t1`.`transaction_id` `거래 식별자`,
    `t1`.`payment_date` `지불 일자`,
    `price` `가격`,
    `quantity` `개수`,
    `item_id` `아이템 이름`
    from
    `tran_1` `t1` left join `tran_d_1` as `td1`
    on `t1`.`transaction_id` = `td1`.`transaction_id`
"""
pd.DataFrame(
    db1.sql_query(query_11)
)
```

```python
# 위의 결과와 같은 작업
# left join 결합
tran_df = pd.merge( tran_1, tran_d_1,
         how= 'left', on= 'transaction_id' )
```

```python
# 정렬 방식
# DataFrame -> sort_values()
# SQL -> order by `컬럼` [ASC: 오름차순, DESC: 내림차순]

query_12 = """
    select * from `emp`
    where `JOB` = "MANAGER"
    order by 'ENAME' asc
"""
# 기본적으로 오름차순 정렬

pd.DataFrame(
    db1.sql_query(query_12)
)
```

**from -> where -> select -> order by 순서로 진행된다.**

```python
# 그룹화 sql
# JOB을 기준으로 그룹화해 SAL의 평균 구하기
# -> JOB과 SAL의 평균을 컬럼으로 지정
# -> SAL의 평균을 기준으로 내림차순 정렬 (order by ... asc/desc)
# -> SAL을 기준으로 그룹화

query_13 = """
    select
    `JOB`, AVG(`SAL`) `mean_SAL`
    from
    `emp`
    group by
    `JOB`
    order by
    `mean_SAL` DESC
"""

pd.DataFrame(
    db1.sql_query(query_13)
)
```
